```python
dir_save = os.path.normpath('../data/dic_20190819.pickle')
dic = load(dir_save)

amt_GI = load_pickle('../data/amt_hGI_pos.pickle')
for i, v in dic.items():
    if i in amt_GI.keys():
        v['AMT'] = amt_GI[i]
amt_ls = list(amt_GI.keys())
ls = amt_ls
```

### reproduce the glove pretrained and nutritions model (i.e. our best model)

```python
presaved_results = load_pickle(pd.read_csv('../csv/2019-08-24 09:17:21.691140.csv').pickle_path[0])
```

```python
# recipe990
original_id = np.array([i for i, v in dic.items() if i in ls])
original_amt = np.array([v['AMT'] for i, v in dic.items() if i in ls])
```

```python
corp0 = fixed_makedata(dic, ls, tag='AMT')
knowns = corp0.knowns
state = 2022
model1 = pretrained_wrap(knowns, 'glove-wiki-gigaword-300')
method = ['wv','nu','scale']
```

```python
corp = copy.deepcopy(corp0)
fold = 0
predictions = []
print('state', state)
ss = StratifiedKFold(n_splits = 5, shuffle = True, random_state = state)
for train_index, test_index in ss.split(corp.ls, corp.y):
    model2 = presaved_results[fold]['model2']
    data = inputs_generater(corp.replace(train_index), model1, method)
    threshold = clf_running_search(data, model1, copy.deepcopy(model2), method)
    data = inputs_generater(corp.add_train_test(train_index, test_index), model1, method)
    y_pred, y_true, name = clf_running(data, model1, copy.deepcopy(model2) ,method, threshold).y_tests
    y_pred = [1 if ele>threshold else 0 for ele in y_pred]
    predictions.append([original_id[test_index], original_amt[test_index], y_pred])
    fold +=1
```

### retrieve the predicted class for each recipe

```python
df_best_answer = pd.concat([pd.DataFrame(predictions[i]).T.rename({0:'id',1:'y_true',2:'y_pred'}, axis = 1) for i in range(5)]).reset_index(drop = True)
df_best_answer['name'] = df_best_answer['id'].apply(lambda x: dic[x]['name'])
df_best_answer['sugar_ratio'] = df_best_answer['id'].apply(lambda x: dic[x]['sugars/dry_weight'])
```

### sugar with very low and very high sugar ratio are easiler to predict

```python
bins = pd.IntervalIndex.from_tuples([(0, 0.05), (0.05, 0.1), (0.1, 0.2), (0.2, 0.3), (0.3, 0.4), (0.4, 0.5),
                                     (0.5, 0.6), (0.6, 0.7), (0.7, 0.8), (0.8, 0.9), (0.9, 1.0),
                                    ])
df_best_answer['sugar_range'] = pd.cut(df_best_answer['sugar_ratio'], bins)
```

```python
for sugar in bins:
    selected = df_best_answer[df_best_answer['sugar_range'] == sugar]
    print(sugar,f1_score(selected['y_true'],selected['y_pred']), len(selected))
```
