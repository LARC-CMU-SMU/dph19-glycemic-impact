```python
import pandas as pd
import numpy as np
```

```python
after_DS = pd.read_csv('data/predict.csv', header=1)
after_DS = after_DS.rename(columns={'Question':'Input.id'})
label_convert = dict(zip([1,2,3,4,5,6],
                         ['1 - Strongly Disagree','2','3','4','5 - Strongly Agree','Not Sure']))
after_DS['Answer.Your answer.label'] = after_DS.label.apply(lambda x: label_convert[int(x[1:])])
after_DS['WorkerId'] = 'Aggregated'
after_DS = after_DS[['Input.id','WorkerId','Answer.Your answer.label']]

amt = pd.read_csv('data/labeled.csv')
amt = amt[['Input.id','WorkerId','Answer.Your answer.label']]

combined = pd.concat([after_DS, amt],axis = 0)
```

```python
combined.to_csv('data/combined.csv',index = False)
```
